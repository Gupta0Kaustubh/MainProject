```python
from pymongo import MongoClient
import snowflake.connector

# Connect to MongoDB
mongo_client = MongoClient('mongodb://localhost:27017/')
mongo_db = mongo_client['your_mongodb_database']
mongo_collection = mongo_db['your_mongodb_collection']

# Connect to Snowflake
snowflake_conn = snowflake.connector.connect(
    user='your_snowflake_username',
    password='your_snowflake_password',
    account='your_snowflake_account',
    warehouse='your_snowflake_warehouse',
    database='your_snowflake_database',
    schema='your_snowflake_schema'
)

# Snowflake table creation SQL
create_table_sql = """
CREATE TABLE IF NOT EXISTS your_snowflake_table (
    field1 datatype,
    field2 datatype,
    -- Add more columns as needed
);
"""

# Insert data into Snowflake
def insert_into_snowflake(doc):
    cur = snowflake_conn.cursor()
    cur.execute("INSERT INTO your_snowflake_table VALUES (?, ?)", (doc['field1'], doc['field2']))
    cur.close()

# Create Snowflake table if it doesn't exist
cur = snowflake_conn.cursor()
cur.execute(create_table_sql)
cur.close()

# Watch for changes in MongoDB
with mongo_collection.watch() as stream:
    for change in stream:
        if change['operationType'] == 'insert':
            # Handle insert operation
            document = change['fullDocument']
            insert_into_snowflake(document)
        elif change['operationType'] == 'update':
            # Handle update operation
            # Note: You may need to implement logic based on your use case
            pass
        elif change['operationType'] == 'delete':
            # Handle delete operation
            # Note: You may need to implement logic based on your use case
            pass

# Close connections
mongo_client.close()
snowflake_conn.close()
```
