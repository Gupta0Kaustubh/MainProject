```python
import random
from faker import Faker
from datetime import timedelta
import csv

# Initialize Faker
fake = Faker()

fake_data_list = []

# Define the list of possible training names
training_names = [
    "Introduction to Python Programming",
    "Web Development Fundamentals",
    "Data Science Essentials",
    "Machine Learning Basics",
    "Introduction to JavaScript",
    "Frontend Development with React",
    "Backend Development with Node.js",
    "Database Management with MongoDB",
    "Cloud Computing Fundamentals",
    "Cybersecurity Essentials"
]

# Generate fake data
for i in range(1, 100):
    certificates = random.sample(training_names, 4)
    start_date = fake.date_between(start_date=datetime(2023, 6, 1), end_date=datetime(2024, 12, 31))
    end_date = start_date + timedelta(days=random.randint(5, 10))  # Random end date 5 to 10 days ahead
    user_id = fake.random_int(min=1, max=200)
    training_data = {
        "trainingId": "Training_" + str(i),
        "trainingName": fake.random_element(elements=certificates),
        "trainingDescription": fake.sentence(),
        "trainerId": "Emp_" + str(user_id),
        "startDate": start_date,
        "endDate": end_date,
        "optimizedDuration": fake.random_int(min=1, max=40),
        "timingOfTraining": f"{start_date.strftime('%Y-%m-%d %H:%M:%S')} - {end_date.strftime('%Y-%m-%d %H:%M:%S')}",  # New column for timing
        "createdAt": fake.date_time_this_year(),
    }
    fake_data_list.append(training_data)

# Write data to CSV file
csv_file_path = "training_data1.csv"
with open(csv_file_path, "w", newline="", encoding="utf-8") as csv_file:
    fieldnames = fake_data_list[0].keys()
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    writer.writeheader()
    for training_data in fake_data_list:
        writer.writerow(training_data)

print(f"Training data has been saved to {csv_file_path}")
```

```python
import csv
import random
from faker import Faker

fake = Faker()

fake_data_list = []

# Define the list of possible quiz names
quiz_names = [
    "Introduction to Python Quiz",
    "Web Development Fundamentals Quiz",
    "Data Science Essentials Quiz",
    "Machine Learning Basics Quiz",
    "Introduction to JavaScript Quiz",
    "Frontend Development with React Quiz",
    "Backend Development with Node.js Quiz",
    "Database Management with MongoDB Quiz",
    "Cloud Computing Fundamentals Quiz",
    "Cybersecurity Essentials Quiz"
]

# Define the list of possible file names for questions
question_files = [
    "questions1.csv",
    "questions2.csv",
    "questions3.csv",
    "questions4.csv",
    "questions5.csv"
]

# Generate fake data
for i in range(1, 100):
    quiz_data = {
        "quizName": fake.random_element(elements=quiz_names),
        "trainingId": "Training_" + str(i),
        "trainingName": fake.sentence(),
        "maxScores": fake.random_int(min=100, max=100),
        "minScores": fake.random_int(min=50, max=50),
        "difficultyLevel": fake.random_element(elements=("easy", "moderate", "hard")),
        "questionFile": fake.random_element(elements=question_files),
    }
    fake_data_list.append(quiz_data)

# Write data to CSV file
csv_file_path = "quiz_data1.csv"
with open(csv_file_path, "w", newline="", encoding="utf-8") as csv_file:
    fieldnames = fake_data_list[0].keys()
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    writer.writeheader()
    for quiz_data in fake_data_list:
        writer.writerow(quiz_data)

print(f"Quiz data has been saved to {csv_file_path}")
```

```python
import csv
import random
from faker import Faker

fake = Faker()

fake_data_list = []

# Define the list of possible training statuses
training_statuses = ['pending', 'in_progress', 'completed']

# Generate fake data
for i in range(1, 1000):
    training_status = fake.random_element(elements=training_statuses)
    if training_status == 'pending':
        assessment_percentage_done = 0
        assessment_completion_time_in_hours= None
    elif training_status == 'completed':
        assessment_percentage_done = 100
        assessment_completion_time_in_hours = fake.random_int(min=1, max=60)
    else:
        assessment_percentage_done = fake.random_int(min=0, max=100)
        assessment_completion_time_in_hours = None
        
    if training_status in ['pending', 'in_progress']:
        score_achieved_in_quiz = None
        quiz_passed_or_failed = None
    else:
        score_achieved_in_quiz = fake.random_int(min=0, max=100)
        if score_achieved_in_quiz < 50:
            quiz_passed_or_failed = 'failed'
        else:
            quiz_passed_or_failed = 'passed'
        
    training_id = fake.random_int(min=1, max=100)
    trainer_id = fake.random_int(min=1, max=100)
    user_id = fake.random_int(min=1, max=200)
    
    tracker_data = {
        "userId": "Emp_" + str(user_id),
        "trainingId": "Training_" + str(training_id),
        "trainingStatus": training_status,
        "assessment_percentage_done": assessment_percentage_done,
        "assessment_completion_time_in_hours": assessment_completion_time_in_hours,
        "scoreAchievedInQuiz": score_achieved_in_quiz,
        "quizPassedOrFailed": quiz_passed_or_failed,
        "ratingGivenByTrainer": fake.random_int(min=1, max=5),
    }
    fake_data_list.append(tracker_data)

# Write data to CSV file
csv_file_path = "tracker_data1.csv"
with open(csv_file_path, "w", newline="", encoding="utf-8") as csv_file:
    fieldnames = ["userId", "trainingId", "trainerId", "trainingStatus", "assessment_percentage_done", 
                  "assessment_completion_time_in_hours", "scoreAchievedInQuiz", "quizPassedOrFailed", 
                  "ratingGivenByTrainer"]
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    writer.writeheader()
    for tracker_data in fake_data_list:
        writer.writerow(tracker_data)

print(f"Tracker data has been saved to {csv_file_path}")
```
